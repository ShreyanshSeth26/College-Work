```python
import torch
from PIL import Image
from transformers import BlipProcessor, BlipForQuestionAnswering
```

```python
image = Image.open("X:/Projects/CV/Assignment 3/human_dog.jpg").convert("RGB")
processor = BlipProcessor.from_pretrained("Salesforce/blip-vqa-base")
model     = BlipForQuestionAnswering.from_pretrained("Salesforce/blip-vqa-base")
model.eval()

question1 = "Where is the dog present in the image?"
inputs1   = processor(image, question1, return_tensors="pt")
out1      = model.generate(**inputs1)
answer1   = processor.decode(out1[0], skip_special_tokens=True)
print("Question 1:", question1)
print("Answer 1:", answer1)

question2 = "Where is the man present in the image?"
inputs2   = processor(image, question2, return_tensors="pt")
out2      = model.generate(**inputs2)
answer2   = processor.decode(out2[0], skip_special_tokens=True)
print("\nQuestion 2:", question2)
print("Answer 2:", answer2)
```

**Comment on VQA outputs**

- **Q1:** “in man’s arms” accurately describes the dog’s location.  
- **Q2:** “living room” correctly identifies the indoor setting.  
- Overall, BLIP successfully grounded both the object location and the scene context.
