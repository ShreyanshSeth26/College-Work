```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader, random_split
import numpy as np
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score
import matplotlib.pyplot as plt
import wandb
import joblib
import random

# Set random seeds for reproducibility
torch.manual_seed(2022484)
np.random.seed(2022484)
random.seed(2022484)
```

```python
# Define A Reusable Function for Running One Experiment
def run_experiment(config):
    wandb.init(project="Fashion_MNIST", name=config["run_name"], config=config)
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f"Using device: {device}")

    # Data loading & transforms
    train_transform = transforms.Compose([
        transforms.RandomHorizontalFlip(),
        transforms.RandomCrop(28, padding=4),
        transforms.ToTensor(),
        transforms.Normalize((0.5,), (0.5,))
    ])
    test_transform = transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize((0.5,), (0.5,))
    ])

    train_val_data = datasets.FashionMNIST(
        root='./data', train=True, download=True, transform=train_transform
    )
    test_data = datasets.FashionMNIST(
        root='./data', train=False, transform=test_transform
    )

    # Split dataset into train and validation
    train_data, val_data = random_split(train_val_data, [50000, 10000])

    # Data loaders
    train_loader = DataLoader(
        train_data, batch_size=config["batch_size"], shuffle=True,
        num_workers=4, pin_memory=True
    )
    val_loader = DataLoader(
        val_data, batch_size=config["batch_size"], shuffle=False,
        num_workers=4, pin_memory=True
    )
    test_loader = DataLoader(
        test_data, batch_size=config["batch_size"], shuffle=False,
        num_workers=4, pin_memory=True
    )

    # Define the Autoencoder
    class FashionAutoEncoder(nn.Module):
        def __init__(self):
            super().__init__()
            # Encoder
            self.encoder = nn.Sequential(
                nn.Conv2d(1, 32, 3, stride=2, padding=1),  # => 14x14
                nn.BatchNorm2d(32),
                nn.ReLU(),
                nn.Conv2d(32, 64, 3, stride=2, padding=1),  # => 7x7
                nn.BatchNorm2d(64),
                nn.ReLU(),
                nn.Flatten(),
                nn.Linear(7*7*64, config["latent_dim"])  
            )
            # Decoder
            self.decoder = nn.Sequential(
                nn.Linear(config["latent_dim"], 7*7*64),
                nn.Unflatten(1, (64, 7, 7)),
                nn.ConvTranspose2d(64, 32, 3, stride=2, padding=1, output_padding=1),
                nn.BatchNorm2d(32),
                nn.ReLU(),
                nn.ConvTranspose2d(32, 1, 3, stride=2, padding=1, output_padding=1),
                nn.Tanh()
            )

        def forward(self, x):
            latent = self.encoder(x)
            reconstructed = self.decoder(latent)
            return reconstructed

    model = FashionAutoEncoder().to(device)
    print(f"Trainable parameters in Autoencoder: {sum(p.numel() for p in model.parameters() if p.requires_grad):,}")

    # Train the Autoencoder
    optimizer = config["optimizer"](model.parameters(), lr=config["learning_rate"])
    scheduler = optim.lr_scheduler.StepLR(
        optimizer,
        step_size=config["scheduler_step"],
        gamma=config["scheduler_gamma"]
    )
    criterion = nn.MSELoss()

    train_losses, val_losses = [], []

    wandb.watch(model, criterion, log="all", log_freq=100)

    for epoch in range(config["epochs_autoencoder"]):
        model.train()
        epoch_loss = 0.0
        for images, _ in train_loader:
            images = images.to(device)
            optimizer.zero_grad()

            outputs = model(images)
            loss = criterion(outputs, images)
            loss.backward()
            optimizer.step()
            epoch_loss += loss.item() * images.size(0)

        train_loss = epoch_loss / len(train_loader.dataset)
        train_losses.append(train_loss)

        # Validation
        model.eval()
        val_epoch_loss = 0.0
        with torch.no_grad():
            for images, _ in val_loader:
                images = images.to(device)
                outputs = model(images)
                val_epoch_loss += criterion(outputs, images).item() * images.size(0)

        val_loss = val_epoch_loss / len(val_loader.dataset)
        val_losses.append(val_loss)

        # Step the LR scheduler
        scheduler.step()
        print(f"[Epoch {epoch+1:02d}] Train Loss: {train_loss:.4f}  |  Val Loss: {val_loss:.4f}")
        
        # Log to wandb
        wandb.log({
            "Auto Encoder Epoch": epoch+1,
            "Auto Encoder Train Loss": train_loss,
            "Auto Encoder Val Loss": val_loss,
            "Auto Encoder Learning Rate": scheduler.get_last_lr()[0]
        })

    # Save the final autoencoder weights
    torch.save(model.state_dict(), f"autoencoder_{config['run_name']}.pth")

    # Plot Autoencoder training curves
    plt.figure()
    plt.plot(train_losses, label='Training Loss')
    plt.plot(val_losses, label='Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('Autoencoder Training Progress')
    plt.legend()
    plt.show()

    # Extract Latent Features
    def extract_features(loader):
        feats, lbls = [], []
        model.eval()
        with torch.no_grad():
            for images, targets in loader:
                images = images.to(device)
                latent = model.encoder(images).cpu().numpy()
                feats.append(latent)
                lbls.append(targets.numpy())
        return np.concatenate(feats), np.concatenate(lbls)

    X_train, y_train = extract_features(train_loader)
    X_val, y_val = extract_features(val_loader)
    X_test, y_test = extract_features(test_loader)

    # Train Multiple SVMs
    svm_models = {
        'Linear SVM (C=0.1)': SVC(kernel='linear', C=0.1),
        'RBF SVM (C=1)': SVC(kernel='rbf', C=1.0),
        'Poly SVM (d=3)': SVC(kernel='poly', degree=3),
        'Sigmoid SVM': SVC(kernel='sigmoid'),
        'RBF SVM (C=10)': SVC(kernel='rbf', C=10)
    }

    best_model_name, best_val_acc = None, 0.0
    for name, svm_model in svm_models.items():
        svm_model.fit(X_train, y_train)
        val_preds = svm_model.predict(X_val)
        val_acc = accuracy_score(y_val, val_preds)
        print(f"{name}: Validation Accuracy = {val_acc:.4f}")
        
        # Log each SVM val accuracy to wandb
        wandb.log({f"{name}_val_acc": val_acc})

        if val_acc > best_val_acc:
            best_val_acc = val_acc
            best_model_name = name

    print(f"\nBest SVM Model on validation set: {best_model_name} with val_acc={best_val_acc:.4f}")

    # Save the best SVM model locally
    best_svm_model = svm_models[best_model_name]
    joblib.dump(best_svm_model, f"best_svm_{config['run_name']}.joblib")

    # MLP Classifier on Latent
    class MLP(nn.Module):
        def __init__(self, input_dim=256):
            super().__init__()
            self.layers = nn.Sequential(
                nn.Linear(input_dim, 128),
                nn.ReLU(),
                nn.Linear(128, 10)
            )

        def forward(self, x):
            return self.layers(x)

    # Adjust input_dim if you changed latent_dim
    mlp = MLP(input_dim=config["latent_dim"]).to(device)
    mlp_optimizer = optim.Adam(mlp.parameters(), lr=config["mlp_lr"])
    mlp_criterion = nn.CrossEntropyLoss()

    train_features = torch.tensor(X_train, dtype=torch.float32).to(device)
    train_labels = torch.tensor(y_train).to(device)
    val_features = torch.tensor(X_val, dtype=torch.float32).to(device)
    val_labels = torch.tensor(y_val).to(device)

    mlp_train_loss, mlp_val_loss = [], []
    mlp_train_acc, mlp_val_acc = [], []

    for epoch in range(config["mlp_epochs"]):
        mlp.train()
        total_loss = 0.0
        correct = 0
        batch_size = config["batch_size"]

        for i in range(0, len(train_features), batch_size):
            batch_features = train_features[i:i+batch_size]
            batch_labels = train_labels[i:i+batch_size]

            mlp_optimizer.zero_grad()
            outputs = mlp(batch_features)
            loss = mlp_criterion(outputs, batch_labels)
            loss.backward()
            mlp_optimizer.step()

            total_loss += loss.item() * batch_labels.size(0)
            preds = outputs.argmax(dim=1)
            correct += (preds == batch_labels).sum().item()

        epoch_train_loss = total_loss / len(train_features)
        epoch_train_acc = correct / len(train_features)
        mlp_train_loss.append(epoch_train_loss)
        mlp_train_acc.append(epoch_train_acc)

        # Validation
        mlp.eval()
        with torch.no_grad():
            val_outputs = mlp(val_features)
            v_loss = mlp_criterion(val_outputs, val_labels)
            val_preds = val_outputs.argmax(dim=1)
            v_acc = (val_preds == val_labels).sum().item() / len(val_labels)

        mlp_val_loss.append(v_loss.item())
        mlp_val_acc.append(v_acc)

        print(f"[MLP Epoch {epoch+1:02d}] Train Loss: {epoch_train_loss:.4f} | Train Acc: {epoch_train_acc:.4f} | Val Loss: {v_loss.item():.4f} | Val Acc: {v_acc:.4f}")
        
        # Log MLP metrics to wandb
        wandb.log({
            "MLP Epoch": epoch+1,
            "MLP Train Loss": epoch_train_loss,
            "MLP Train Accuracy": epoch_train_acc,
            "MLP Val Loss": v_loss.item(),
            "MLP Val Accuracy": v_acc
        })

    # Save the MLP weights locally
    torch.save(mlp.state_dict(), f"mlp_{config['run_name']}.pth")

    # Final Evaluation on Test
    test_preds_svm = best_svm_model.predict(X_test)

    mlp.eval()
    with torch.no_grad():
        test_features_t = torch.tensor(X_test, dtype=torch.float32).to(device)
        test_preds_mlp = mlp(test_features_t).argmax(1).cpu().numpy()

    print("\nBest SVM Classification Report on Test Set:")
    print(classification_report(y_test, test_preds_svm))

    print("MLP Classification Report on Test Set:")
    print(classification_report(y_test, test_preds_mlp))

    # Plot MLP training curves
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

    ax1.plot(mlp_train_loss, label='Train Loss')
    ax1.plot(mlp_val_loss, label='Val Loss')
    ax1.set_title('MLP Loss Curve')
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('Loss')
    ax1.legend()

    ax2.plot(mlp_train_acc, label='Train Accuracy')
    ax2.plot(mlp_val_acc, label='Val Accuracy')
    ax2.set_title('MLP Accuracy Curve')
    ax2.set_xlabel('Epoch')
    ax2.set_ylabel('Accuracy')
    ax2.legend()
    plt.show()
    wandb.finish()

# Run the Experiment 3 Times with Different Configs
if __name__ == "__main__":
    # 3 distinct config sets
    config_list = [
        {
            "run_name": "EXP1",
            "epochs_autoencoder": 15,
            "batch_size": 256,
            "learning_rate": 0.001,
            "scheduler_step": 5,
            "scheduler_gamma": 0.5,
            "latent_dim": 256,    # AE bottleneck dimension
            "optimizer": optim.Adam,
            "mlp_epochs": 15,
            "mlp_lr": 0.001
        },
        {
            "run_name": "EXP2",
            "epochs_autoencoder": 15,
            "batch_size": 256,
            "learning_rate": 0.0005,
            "scheduler_step": 7,
            "scheduler_gamma": 0.3,
            "latent_dim": 128,    # smaller AE bottleneck
            "optimizer": optim.Adam,
            "mlp_epochs": 15,
            "mlp_lr": 0.0005
        },
        {
            "run_name": "EXP3",
            "epochs_autoencoder": 15,
            "batch_size": 128,
            "learning_rate": 0.01,  # using SGD as the optimizer here
            "scheduler_step": 5,
            "scheduler_gamma": 0.5,
            "latent_dim": 64,     # even smaller AE bottleneck
            "optimizer": optim.SGD,
            "mlp_epochs": 15,
            "mlp_lr": 0.01
        }
    ]

    # Run each experiment in turn
    for cfg in config_list:
        run_experiment(cfg)
```

## Autoencoder Design and Model Summary
I designed a custom autoencoder using PyTorch to learn latent features from the Fashion MNIST dataset. My autoencoder consists of an encoder—comprising two convolutional layers with batch normalization and ReLU activations, followed by a fully connected layer that compresses the input into a low-dimensional (bottleneck) representation—and a decoder that mirrors this architecture using transposed convolutions to reconstruct the original image. I report the number of trainable parameters (e.g., 1,646,849 in one configuration) to illustrate my model’s capacity and complexity.

## Data Augmentation Techniques
To improve generalization and reduce overfitting, I applied data augmentation to the training set. Specifically, I used random horizontal flipping and random cropping. These augmentations increased the diversity of my training data, enabling the autoencoder to learn more robust features that are less sensitive to variations in the input images.
